```python
import time

def calc_fib(k):
    if (k == 0):
        return 1
    
    return calc_fib(k - 1) + calc_fib(k - 1)

def calc_fib_faster(k):
    if (k == 0):
        return 1
    
    return 2 * calc_fib_faster(k - 1)

# print(calc_fib(0))
# print(calc_fib(1))
# print(calc_fib(2))
# print(calc_fib(3))
# print(calc_fib(4))

max_fib = 27

for i in range(max_fib):
    start_time = time.time()
    fiber = calc_fib(i)
    baseline_time = time.time() - start_time

    start_time = time.time()
    fiber = calc_fib_faster(i)
    faster_time = time.time() - start_time

    print(i, baseline_time, faster_time)
```
